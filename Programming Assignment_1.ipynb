### 2. Write a Python program to do arithmetical operations addition and division.?

```python
# takes two numbers and an operator as input from the user. Later the operation is performed

#Here numbers and operator is taken as input
num1 = int(input('Enter Number 1: '))
num2 = int(input('Enter Number 2: '))
op = input('Enter operation to perform: ')

#Performing operations. An else statement is used for error handling
if op == '+':
    print('\nThe answer for given equation is: ', num1 + num2)
elif op == '/':
    print('\nThe answer for given equation is: ', num1 / num2)
else:
    print('\nThe given operation is not supported. Try again!')
```

### 3. Write a Python program to find the area of a triangle?

```python
# take length of the base and height as input
base = float(input('Enter the length of the base of a triangle: '))
height = float(input('Enter the length of the height of a triangle: '))

#ouput the area using the formula
print('\nThe area of the traingle is: ', 0.5 * base * height, 'sq.units')
```

### 4. Write a Python program to swap two variables?

```python
# take two variables and input values in them from user
var1 = input('Enter data into var1: ')
var2 = input('Enter data into var2: ')

# store var1 value in temp_var, store var2 value in var1 and store temp_var in var2
temp_var = var1
var1 = var2
var2 = temp_var

#output the swapped variables
print('\nVALUES AFTER SWAPPING: ')
print('var1: ', var1)
print('var2: ', var2)
```

### 5. Write a Python program to generate a random number?

```python
# import random library
import random

#take upper and lower limit from the user
lower = int(input('Enter the lower limit for random variable: '))
upper = int(input('Enter the upper limit for random variable: '))

#print the random variable using randint() function
print('\nThe random variable is: ', random.randint(lower, upper))
```
